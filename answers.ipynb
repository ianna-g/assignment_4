```python
#elevation and rock cover relationship

from matplotlib import pyplot as plt
import seaborn as sns

y = env['COVER_ROCK']
x = env['ELEVATION']
plt.title("Relationship between Elevation and Rock Cover") 

plt.scatter(x,y) 
#regression line
sns.regplot(x, y, ci=None, line_kws={'color': 'red'})
plt.xlabel("Elevation") 
plt.ylabel("Cover Rock percentage") 
plt.show()

#The relationship is not positive. I added a regression line to this plot(in red) and it is visible that the line is increasing but is so
#minor that it can't be considered a positive relationship.


#distribution of the totals

t = species['totals']
plt.hist(t)
plt.xlabel("Row Number") 
plt.ylabel("Total Values") 
plt.title("Distribution of the Totals") 
plt.show()
print(species['totals'].describe) #summary statistics


#subset of env

sub_names = ['Caricetum scirpoideae-rupestris', 'Pedicularo kanei-Caricetum glacialis', 'Saxifrago tricuspidatae-Artemisietum alaskanae']
fig = env.loc[(env['PLANT_COMMUNITY_NAME'] == sub_names[0])| (env['PLANT_COMMUNITY_NAME'] == sub_names[1])|(env['PLANT_COMMUNITY_NAME'] == sub_names[2]),:]

fig = fig.replace(sub_names[0], 'Caricetum')
fig = fig.replace(sub_names[1], 'Pedicularo')
fig = fig.replace(sub_names[2], 'Saxifrago')

boxplot_df = fig.reset_index(drop=True)
# env[(env['PLANT_COMMUNITY_NAME'] == cols[0]),:]

sns.boxplot(data=boxplot_df, y="COVER_BARE_SOIL", x="PLANT_COMMUNITY_NAME")
plt.xlabel("Plant Community Name") 
plt.ylabel("Bare Soil Cover") 
plt.title("Box-plot for species and bare soil cover") 
plt.show()

#There are no differences becauese the boxes, median line overlap in that one box is not above the other and in a way align(are on teh same line). 
#None of the median lines are outside of the boxes of the two other species samples. 
# The highest median value is with the plant community Saxifrago tricuspidatae-Artemisietum alaskanae and the lowest median
#value is with the Pedicularo kanei-Caricetum glacialis plant family. Caricetum scirpoideae-rupestris has the most spread and
# Pedicularo kanei-Caricetum glacialis has the least spread.


```
